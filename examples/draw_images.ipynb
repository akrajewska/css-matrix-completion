```python
import pandas as pd
from pathlib import Path

OUTPUT_DIR = '/home/tosia/NASK/css-matrix-completion/examples/OUTPUT'
output_file = 'output_im'

mapping_alg = {180: "CSNN_0.5", 216: "CSNN_0.6", 251: "CSNN_0.7", 360: "NN"}

columns =  ['trial', 'missing_part', 'n_cols', 'elapsed_time', 'rmse', 'rmse_omega', 'mae']
df = pd.read_csv(Path(OUTPUT_DIR) / output_file, names=columns)
df= df[(df.n_cols != -1)]
df["algorithm"] = df.n_cols.map(mapping_alg)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('whitegrid')
cmap = sns.color_palette("RdBu", 4)
sns.set_palette(cmap)
ax = sns.barplot(x='missing_part', y='rmse_omega', hue='algorithm',
                  data=df)
# ax.set(yscale='log')
ax.set_ylabel(r'$\mathrm{RMSE } {\Omega}$')
ax.set_xlabel(r'missing part')
ax.set_ylim(0, 1)
plt.savefig('rmse_images')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('whitegrid')
cmap = sns.color_palette("RdBu", 4)
sns.set_palette(cmap)
ax = sns.barplot(x='missing_part', y='elapsed_time', hue='algorithm',
                  data=df)
# ax.set(yscale='log')
ax.set_ylabel(r'$time (s)$')
ax.set_ylim(0, 150)
ax.set_xlabel(r'missing part')
plt.savefig('time_images')
```
