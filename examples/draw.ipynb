```python
import pandas as pd
from pathlib import Path

OUTPUT_DIR = '/home/tosia/NASK/css-matrix-completion/examples/OUTPUT'
output_file = 'output'

columns =  ['trial', 'n_rows', 'n_cols', 'rank', 'fraction_missing', 'n_selected_cols', 'elapsed_time', 'rmse', 'rmse_omega', 'mae']
df = pd.read_csv(Path(OUTPUT_DIR) / output_file, names=columns)

mapping_alg = {100: "CSNN_0.2", 250: "CSNN_0.5", 350: "CSNN_0.7", 500: "NN"}
df["algorithm"] = df.n_selected_cols.map(mapping_alg)
df = df[(df["rank"]==5)]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('whitegrid')
cmap = sns.color_palette("RdBu", 4)
sns.set_palette(cmap)

ax = sns.barplot(x='fraction_missing', y='rmse_omega', hue='algorithm',
                  data=df)
ax.set(yscale='log')
ax.set_ylabel(r'$\mathrm{RMSE} {\Omega}$')

plt.savefig('rmse_synthetic_missing_data')
```
